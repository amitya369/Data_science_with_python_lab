```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2, 2, figsize=(10, 8), dpi=100)
for i, ax in enumerate(axes.flat, start=1):
     ax.plot(range(10), [x * i for x in range(10)])
     ax.set_title(f'Subplot {i}')
plt.tight_layout()
plt.show()

# AMIT KUMAR YADAV
```

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.normal(size=1000)
plt.figure()
plt.scatter(np.arange(len(data)), data, color="blue", alpha=0.5)
plt.title("Scatter Plot")
plt.show()
plt.figure()
plt.hist(data, bins=30, color="green", alpha=0.7)
plt.title("Histogram")
plt.show()
plt.figure()
plt.boxplot(data)
plt.title("Boxplot")
plt.show()

# AMIT KUMAR YADAV
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.1, 100, 100)
y = np.log(x)
plt.figure()
plt.plot(x, y)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Logarithmic X-axis")
plt.ylabel("Logarithmic Y-axis")
plt.title("Logarithmic Scale Plot")
plt.show()

# AMIT KUMAR YADAV
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = np.random.normal(size=100)
df = pd.DataFrame({"x": data, "y": data + np.random.normal(size=100)})
sns.histplot(data, kde=True)
plt.show()
sns.jointplot(x="x", y="y", data=df)
plt.show()
sns.pairplot(df)
plt.show()
sns.rugplot(data)
plt.show()

# AMIT KUMAR YADAV
```
